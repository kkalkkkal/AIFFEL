```python
#검색결과 url 만드는 함수
def insta_searching(word):
    url = 'https://www.instagram.com/explore/tags/' + word
    return url
```

```python
#검색결과 페이지 접속하기
word = '제주도맛집'
url = insta_searching(word)
driver.get(url)
```

```python
#첫 번째 게시글 열기
def select_first(driver):
    first = driver.find_element_by_css_selector('div._9AhH0')
    first.click()
    time.sleep(3)
select_first(driver)
```

```python
#게시글 정보 가져오기
import re
from bs4 import BeautifulSoup
import unicodedata

def get_content(driver):
    # ① 현재 페이지 html 정보 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    try:
        content = soup.select('div.MOdxS')[0].text #위치값 수정 -https://mokeya.tistory.com/68
        content = unicodedata.normalize('NFC', content) 
    except:
        content = ' '
    # ③ 본문 내용에서 해시태그 가져오기(정규식 활용)
    tags = re.findall(r'#[^\s#,\\]+', content)  
    # ④ 작성일자 정보 가져오기
    date = soup.select('time._1o9PC')[0]['datetime'][:10] #위치값 클래스 명 수정 완료
    # ⑤ 좋아요 수 가져오기
    try:
        like = soup.select('section.EDfFK.ygqzn')[0].findall('span')[-1].text #다른 코드 참조   
    except:
        like = 0
    # ⑥ 위치정보 가져오기
    try: 
        place = soup.select('div.M30cS')[0].text
        place = unicodedata.normalize('NFC', place)
    except:
        place = ''
    # ⑦ 수집한 정보 저장하기
    data = [content, date, like, place, tags]
    return data

get_content(driver)
```

```python
#다음 게시글 열기(옆 화살표 누르는 함수)
def move_next(driver):
    right = driver.find_element_by_css_selector ("div.l8mY4.feth3 > .wpO6b")
    right.click()
    time.sleep(3)

move_next(driver)

#해결 레퍼런스
# https://github.com/SOMJANG/Instagram_Crawler/issues/4
```

```python
### 인스타그램 검색페이지 URL 만들기 ###
word = "제주도맛집"    #검색어
url = insta_searching(word)

# ③ 검색페이지 접속하기
driver.get(url)
time.sleep(5)

# ④ 첫 번째 게시글 열기
select_first(driver)

# ⑤ 비어있는 변수(results)만들기
results = [ ]


# ⑥→⑦→⑧ 여러 게시물 수집하기
target = 50      # 크롤링할 게시글 수
for i in range(target):
    # 게시글 수집에 오류 발생시(네트워크 문제 등의 이유로)  2초 대기 후, 다음 게시글로 넘어가도록 try, except 구문 활용
    try:
        time.sleep(5)
        data = get_content(driver)    # 게시글 정보 가져오기
        results.append(data)
        move_next(driver)
    except:
        time.sleep(2)
        move_next(driver)
    
```

```python
# check sample
print(results)
```

```python
# 예제 5-11 크롤링 결과 저장하기

import pandas as pd

results_df = pd.DataFrame(results)
results_df.columns = ['content','data','like','place','tags']
results_df.to_excel('./week_04_1_crawling_jejudoMatJip.xlsx')
```
