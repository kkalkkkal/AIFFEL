### ※ `URL 구조 이해하기`

url 구성 : 주소 `?` 파라미터
    - 주소 
        - 주소ex): https://search.naver.com/search.naver
        - 주소ex): http://search.daum.net/search

    - 파라미터: 변수명1=변수값1 & 변수명2=변수값2....    
        - 주소 뒤에 ?  +   변수명=값  으로 표시. 변수사이에는 & 로 구분
     

※ url 에서는 한글 사용 x : 한글입력시 --> 자동 전환

url을 추정하고 변수가 무엇인지 잘 만지면서 찾아야한다.

```python
word_list = ['모두의연구소', '파이썬', '부동산']
for i in word_list : 
    word = i
#word = '모두의연구소'
    url = f'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query={word}' # 앞에 f를 붙여야 넣을 수 있다.
    print(url)
```

## HTML 살펴보기

---
1. 모든 태그는 태그명 을 가짐  
    `<태그명>`
---
2. 시작과 끝이 있음  
  `<태그명>  </태그명>`
---
3. 태그 내에 속성이 사용될 수 있음  
    `<태그명 속성1=값1> </태그명>`  
    `<태그명 속성1=값1 속성2=값2> </태그명>`  
---
4. 태그는 다른 태그에 포함될 수 있음

    <부모태그>
        <자식태그> </자식태그>
    </부모태그>  
---  
5. 태그 기호 사이에 화면에 보이는 텍스트가 포함되어 있음
  
  `<태그> 안녕하세요</태그>`
---

# 유튜브 랭크 크롤링


```python
# URL 확인하기
url = "https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page=1"
print(url)
```

```python
import time    #  time.sleep(대기시간)
import pandas as pd   
```

```python
# 페이지 접속하기
browser.get(url)
```

```python
html = browser.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 크롬브라우저 -> 개발자도구 활용시   <tr> 태그에 채널별 정보뭉치(총 100개)가 있는 것 확인 가능

channel_list = soup.select('tr')   # soup.select('tr') ->  " soup 에서 태그명이 tr 인 거 다 찾아줘~!"  
len(channel_list)                  # 102 ->   찾고 있는 100개 외에 2개의 다른 정보들도 포함되어 있음
```

```python
# 100개가 나올때까지 정보를 하나씩 추가해가며 찾습니다. 
channel_list = soup.select('tbody > tr')  #  "soup 에서 tbody 바로 아래에 있는 tr 태그 다 찾아줘"
len(channel_list)
```

```python
# 100개가 나올때까지 정보를 하나씩 추가해가며 찾습니다. 
channel_list = soup.select('table > tbody > tr')  #  "soup 에서 table 바로 아래의, tbody 바로 아래에 있는 tr 태그 다 찾아줘"
len(channel_list)
```

```python
# 100개가 나올때까지 정보를 하나씩 추가해가며 찾습니다. 

#  "soup 에서 form 바로 아래 table 바로 아래의, tbody 바로 아래에 있는 tr 태그 다 찾아줘"
channel_list = soup.select('form > table > tbody > tr')  
len(channel_list)   # 100 !!!  원하는 채널별 정보 뭉치를 딱 찾았습니다 !! 
```

```python
# 찾은 데이터들을 하나씩 살펴봅니다. 
channel_list[0]
```

# (추가)여러페이지 가져오기

```python
# URL 에서 페이지 정보 변경 가능(문자열 포매팅 활용하기!)
page = 2

url = f"https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page={page}"
print(url)
```

- 여러 페이지 / 다수의 반복작업 진행시 -> 진행 상태바 추가하기

```python
results = []
```

```python
for page in range(1,11) : 
    url = f"https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page={page}"
    browser.get(url)
    time.sleep(1)
    soup = BeautifulSoup(browser.page_source, "html.parser")
    channel_list = soup.select('form > table > tbody > tr')
    
    for channel in channel_list : 
        title = channel.select('h1 > a')[0].text.strip()
        category = channel.select('h1 > p')[0].text.strip()
        subscriber = channel.select('td.subscriber_cnt')[0].text
        view = channel.select('td.view_cnt')[0].text
        video = channel.select('td.video_cnt')[0].text
        print(title, category, subscriber, view, video)
        data = [title, category, subscriber, view, video]
        results.append(data)
```

```python
df = pd.DataFrame(results)
df.columns = ['채널명', '카테고리','구독자수','조회수','영상수']
df.to_excel('./유튜브채널정보.xlsx', index = False)
```
