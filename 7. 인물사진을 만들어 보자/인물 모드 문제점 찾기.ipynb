```python
from PIL import Image
import glob
```

```python
# resize 

def resize_image(img_path):
    # 파일마다 모두 700x900 사이즈로 바꾸어 저장합니다.
    target_size=(700,900)

    old_img=Image.open(img_path)
    new_img=old_img.resize(target_size,Image.ANTIALIAS)
    new_img = new_img.convert("RGB") #png일 경우 alpha 값 때문에 에러남.
    new_img.save(img_path, "JPEG")
    
    print(img_path, "image resized.")

    
# 각 이미지가 저장된 디렉토리 아래의 모든 jpg 파일을 읽어들여서 resize

resize_image(img_path)
resize_image(background_path)
resize_image(cat_path)


print("모든 이미지 resize 완료!")
```

```python
#이미지 경로 준비하기
img_path = './data/my_face.jpg'  
cat_path = './data/my_cat.jpg'
background_path = './data/my_background.jpg'

# 이미지 불러오기
img_resized = cv2.imread(img_path) 
cat_resized = cv2.imread(cat_path)
background_resized = cv2.imread(background_path)

print('인물 사진 크기 : ', img_resized.shape)
print('고양이 사진 크기 : ', cat_resized.shape)
print('배경 사진 크기 : ', background_resized.shape)
```

```python
# Pixlib에서 제공하는 모델 다운로드 
# 저장할 파일 이름을 결정합니다
model_dir = './model'
model_file = os.path.join(model_dir, 'deeplabv3_xception_tf_dim_ordering_tf_kernels.h5')

# PixelLib가 제공하는 모델의 url입니다
model_url = 'https://github.com/ayoolaolafenwa/PixelLib/releases/download/1.1/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5'

# 다운로드를 시작합니다
urllib.request.urlretrieve(model_url, model_file)
```

# Step 1. 인물모드 직접 해 보기

### 세그멘테이션으로 사람 분리하기

```python
#세그멘테이션 모델 생성
model = semantic_segmentation()
model.load_pascalvoc_model(model_file)
```

```python
# 모델 이미지 입력 (인물, 고양이)
man_segvalues, man_output = model.segmentAsPascalvoc(img_path)
cat_segvalues, cat_output = model.segmentAsPascalvoc(cat_path)
```

### 세그멘테이션 뚜렷하게 바꾸기

```python
colormap = np.zeros((256, 3), dtype = int)
ind = np.arange(256, dtype=int)

for shift in reversed(range(8)):
    for channel in range(3):
        colormap[:, channel] |= ((ind >> channel) & 1) << shift
    ind >>= 3

colormap[:20]
```

```python
# 사람, 고양이, 배경에 대한 구분색 확인하기
print(colormap[15]) # 사람
print(colormap[8]) # 고양이
print(colormap[0]) # 배경
```

```python
seg_color1 = (128,128,192) # 사람
seg_color2 = (0,0,64) # 고양이
seg_color3 = (0,0,0) # 배경
```

```python
# 인물 사진
seg_map = np.all(man_output==seg_color1, axis=-1)
print(seg_map.shape)
plt.imshow(seg_map, cmap='gray')
plt.show()
```

```python
#고양이 사진
seg_map2 = np.all(cat_output==seg_color2, axis=-1)
print(seg_map2.shape)
plt.imshow(seg_map2, cmap='gray')
plt.show()
```

### 원래 이미지와 겹쳐 확인하기



```python
# 인물사진 

img_show = img_resized.copy()
img_mask = seg_map.astype(np.uint8) * 255
color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)
img_show = cv2.addWeighted(img_show, 0.6, color_mask, 0.4, 0.0)

plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

머리카락 부분이 잘 구분되지 못한 것으로 보인다. 

```python
# 고양이 사진 

img_show2 = cat_resized.copy()
img_mask2 = seg_map2.astype(np.uint8) * 255
color_mask2 = cv2.applyColorMap(img_mask2, cv2.COLORMAP_JET)
img_show2 = cv2.addWeighted(img_show2, 0.6, color_mask2, 0.4, 0.0)

plt.imshow(cv2.cvtColor(img_show2, cv2.COLOR_BGR2RGB))
plt.show()
```

고양이는 잘 구분되었으나, 주변 배경이 완벽하게 구분되지 않은 것으로 보인다.

## 배경 흐리게 하기

### 1. 인물 사진

#### 블러 처리

```python
img_resized_blur = cv2.blur(img_resized, (13,13))  
plt.imshow(cv2.cvtColor(img_resized_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 마스크를 씌워 배경만 분리하기

```python
img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)
img_bg_mask = cv2.bitwise_not(img_mask_color)
img_bg_blur = cv2.bitwise_and(img_resized_blur, img_bg_mask)
plt.imshow(cv2.cvtColor(img_bg_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 흐린 배경 합성하기

```python
img_concat = np.where(img_mask_color==255, img_resized, img_bg_blur)
plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
plt.show()
```

### 2. 고양이 사진

#### 블러 처리

```python
cat_resized_blur = cv2.blur(cat_resized, (13,13))  
plt.imshow(cv2.cvtColor(cat_resized_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 마스크를 씌워 배경만 분리하기

```python
cat_mask_color = cv2.cvtColor(img_mask2, cv2.COLOR_GRAY2BGR)
cat_bg_mask = cv2.bitwise_not(cat_mask_color)
cat_bg_blur = cv2.bitwise_and(cat_resized_blur, cat_bg_mask)
plt.imshow(cv2.cvtColor(cat_bg_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 흐린 배경 합성하기

```python
cat_concat = np.where(cat_mask_color==255, cat_resized, cat_bg_blur)
plt.imshow(cv2.cvtColor(cat_concat, cv2.COLOR_BGR2RGB))
plt.show()
```

이렇게 두 사진에 대해 아웃포커싱 효과를 적용시켜보았다.

# Step 2. 사진에서 문제점 찾기

인물 사진의 경우 사람과 배경은 잘 구분해주었으나, 보다시피 삐져나온 머리카락 부분의 누끼가 잘 안 떠졌다. 

고양이 사진의 경운 고양이 부분의 누끼는 잘 따졌으나, 배경에 대한 구분이 완벽하게 잘 되지 않은 것으로 보인다. <br>
고양이 누끼를 기준으로 나머지를 배경으로 재지정하면 되기 때문에 큰 문제가 아니다.

# Step 3. 해결 방법을 제안해 보기

### 1. 누끼 경계값에서 인접 픽셀 찾아가기  

머리카락의 경우 누끼 경계 값에서 유사한 RGB를 가지고 있을 경우 누끼를 하나하나 확장하는 방법이 있다. <br>

알고리즘 문제로 치면 DFS/BFS 문제인데,예시 이미지의 경우 배경과 머리카락의 RGB 값이 극명하게 차이나서 쉽게 머리카락을 구분할 수 있겠으나 배경과 구분이 어려울 경우 다른 방법을 더 고안해보아야 할 것이다. 

누끼 경계라인에서 3x3 크기의 탐색을 진행하고 색상값이 일치하면 깊이 탐색을 확장하는 알고리즘을 사용하면 될 것이다. 그럼 이어진 머리카락을 하나하나 누끼에 포함시킬 수 있을 것이다.
